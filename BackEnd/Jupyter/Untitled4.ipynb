```python
def api_def(args):
    response = {}
    try:
        string = args.get('string')
        reversed_s = string[::-1]
        response['result'] = reversed_s
        response['status'] = 200
    except Exception as e:
        response['status'] = 404
        response['error'] = str(e)
    return response
```

```python
api_def({'string' : 'Rathan'})
```
