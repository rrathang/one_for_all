```python
def api_def(args):
    exec("import requests", globals())
    exec("import json", globals())
    response = {}
    try:
        message = args.get('message')
        headers = {
        'Content-Type': 'application/json'
        }
        payload = {
            "type": "message",
            "attachments": [
                {
                    "contentType": "application/vnd.microsoft.card.adaptive",
                    "contentUrl": "",
                    "content": {
                        "$schema": "",
                        "type": "AdaptiveCard",
                        "version": "1.2",
                        "body": [
                            {
                                "type": "TextBlock",
                                "text": f"{message}",
                                "weight": "bolder",
                                "size": "medium",
                                "width": "stretch"
                            }
                        ]
                    }
                }
            ]
        }
        webhook_url = 'https://prod-130.westeurope.logic.azure.com:443/workflows/06917a5af5194e79ac35b9d15b103241/triggers/manual/paths/invoke?api-version=2016-06-01&sp=%2Ftriggers%2Fmanual%2Frun&sv=1.0&sig=JjAUDqQNEsRqSFIgFmEFd-zF7zhm3TfzZ5Ve4sriRxc' 
        webhook_response = requests.post(webhook_url, headers=headers, data=json.dumps(payload))
    
        if webhook_response.status_code == 202:
            response['status'] = 202
            response['result'] = 'Message sent successfully'
        else:
            response['status'] = webhook_response.status_code
            response['result'] = f'Failed to send message. Status code: {response.status_code}, Response: {response.text}'
            
    except Exception as e:
        response['status'] = 404
        response['error'] = str(e)
    return response
```
