```python
def api_def(*args, **kwargs):
    exec("import requests", globals())
    exec("import json", globals())

    response = {}
    try:
        # Extract the message and dictionary from args
        message = kwargs.get('message', '')
        table_data = kwargs.get('table_data', {})

        # Prepare the headers
        headers = {
            'Content-Type': 'application/json'
        }

        # Create the payload for the adaptive card
        card_body = [
            {
                "type": "TextBlock",
                "text": f"{message}",
                "weight": "bolder",
                "size": "medium",
                "width": "stretch"
            }
        ]

        # Check if table_data is a dictionary and add it as rows
        if isinstance(table_data, dict) and table_data:
            # Header row for the table
#             card_body.append({
#                 "type": "ColumnSet",
#                 "columns": [
#                     {"type": "Column", "width": "auto", "items": [{"type": "TextBlock", "text": "Attribute_Name", "weight": "bolder"}]},
#                     {"type": "Column", "width": "auto", "items": [{"type": "TextBlock", "text": "Attribute_Value", "weight": "bolder"}]}
#                 ]
#             })

            # Add each key-value pair as a row in the table
            for key, value in table_data.items():
                card_body.append({
                    "type": "ColumnSet",
                    "columns": [
                        {"type": "Column", "width": "auto", "items": [{"type": "TextBlock", "text": str(key) + " :"}]},
                        {"type": "Column", "width": "auto", "items": [{"type": "TextBlock", "text": str(value)}]}
                    ]
                })

        # Construct the adaptive card payload
        payload = {
            "type": "message",
            "attachments": [
                {
                    "contentType": "application/vnd.microsoft.card.adaptive",
                    "contentUrl": "",
                    "content": {
                        "$schema": "http://adaptivecards.io/schemas/adaptive-card.json",
                        "type": "AdaptiveCard",
                        "version": "1.2",
                        "body": card_body
                    }
                }
            ]
        }

        # Set the webhook URL
        webhook_url = 'https://prod-130.westeurope.logic.azure.com:443/workflows/06917a5af5194e79ac35b9d15b103241/triggers/manual/paths/invoke?api-version=2016-06-01&sp=%2Ftriggers%2Fmanual%2Frun&sv=1.0&sig=JjAUDqQNEsRqSFIgFmEFd-zF7zhm3TfzZ5Ve4sriRxc'  # replace with the actual URL

        # Send the request to the webhook
        webhook_response = requests.post(webhook_url, headers=headers, data=json.dumps(payload))

        # Check response status
        if webhook_response.status_code == 202:
            response['status'] = 202
            response['result'] = 'Message sent successfully'
        else:
            response['status'] = webhook_response.status_code
            response['result'] = f"Failed to send message. Status code: {webhook_response.status_code}, Response: {webhook_response.text}"
    
    except Exception as e:
        response['status'] = 404
        response['error'] = str(e)
    
    return response
```
